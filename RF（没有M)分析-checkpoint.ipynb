```python
import pandas as pd
```

```python
df1 = pd.read_csv('D:/数据科学/淘宝数据/SQL/用户F值.csv',encoding='gbk')
df1
```

```python
df2 = pd.read_csv('D:/数据科学/淘宝数据/SQL/用户R值.csv',encoding='gbk')
df2
```

```python
df = pd.merge(df1,df2,on='用户ID')
def 用户层级(f,r):
    if r>3 and f>1:
        return '重要价值客户'
    if r>3 and f<1:
        return '重要发展客户'
    if r<3 and f>1:
        return '重要保留客户'
    else: return '重要挽留客户'
    
df['用户层级']=df.apply(lambda x:用户层级(x.F值,x.R值),axis=1)
```
